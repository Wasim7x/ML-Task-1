```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('temperature.csv')
df
```

```python
type(df['Country'])
```

```python
df = df[['Region','Country','Year','AvgTemperature']]
df.head
```

world avg temp

```python
df1 = df[['Year','AvgTemperature']]
```

```python
df1 = df1.groupby('Year')['AvgTemperature'].median().reset_index()
df1
```

```python

df1 = df1.iloc[2:-1, : ]
df1['Change_Temp'] = df1['AvgTemperature'] - df1['AvgTemperature'].iloc[0]
```

```python
plt.plot(df1['Year'], df1['Change_Temp'], label='CO2', color='g')
plt.xlabel("Year")
plt.ylabel('Change_Temp')
plt.title('Change_Temp')
plt.grid(True)
plt.legend()
plt.show()
```

CONTINANATE AVG TEMP

```python
df2 = df[['Region','AvgTemperature']]
df2
```

```python
min_temp = df2.groupby('Region')['AvgTemperature'].min().reset_index()
min_temp
```

```python
max_temp = df2.groupby('Region')['AvgTemperature'].max().reset_index()
max_temp
```

```python
change_temp = max_temp['AvgTemperature'] - min_temp['AvgTemperature']
change_temp
```

```python
plt.bar(max_temp['Region'], change_temp) 
plt.xlabel('Region')
plt.ylabel('Average Temperature (Â°F)')
plt.title('Change in Temprature (Â°F)')
plt.xticks(rotation=45 )
plt.show()
```
