```python
import pandas as pd
import numpy as nu
import matplotlib.pyplot as plt
df = pd.read_csv('co2.csv', encoding='latin1')
```

```python
df
```

```python
df["Year"].dtype
```

```python
df['co2 emition per person(tons)']=df['CO2 emission (Tons)']/df['Population(2022)']
df['co2 emition per person(tons)']
```

```python
df
```

```python
p = df[df['CO2 emission (Tons)'] != 0]
```

```python
mini = p['co2 emition per person(tons)'].min()
mini
```

```python
df[df['co2 emition per person(tons)']==mini]['Country']
```

co2 emission in countries in econimicaly tranition


```python
df2 = df.groupby('Year').agg({"co2 emition per person(tons)": "median"}).reset_index()
df3 = df[df['Country'] == 'India'].groupby('Year').agg({"co2 emition per person(tons)": "median"})
df4 = df[df['Country'] == 'Turkey'].groupby('Year').agg({"co2 emition per person(tons)": "median"})
df5 = df[df['Country'] == 'Brazil'].groupby('Year').agg({"co2 emition per person(tons)": "median"})
df2_filtered = df2[df2['Year'] > 1900]
df2_filtered
```

```python
df2_filtered
```

```python
plt.plot(df2_filtered['Year'], df2_filtered['co2 emition per person(tons)'], label='world', color='r')
plt.plot(df3['co2 emition per person(tons)'].keys(), df3['co2 emition per person(tons)'].values,label= 'india',color='g')
plt.plot(df4['co2 emition per person(tons)'].keys(), df4['co2 emition per person(tons)'].values,label= 'Turkey',color='b')
plt.plot(df5['co2 emition per person(tons)'].keys(), df5['co2 emition per person(tons)'].values,label= 'Brazil',color='y')
plt.xlabel("Year")
plt.ylabel('co2 emission in countries in econimicaly tranition')
plt.title('World Avg Carbon Emission in eqnomicaly modrate')
plt.grid(True)
plt.legend()
plt.show()
```
